```python
import hashlib

import pandas as pd
```

```python
def md5_hash(val):
    return hashlib.md5(str(val).encode('utf-8')).hexdigest()
```

```python
events = pd.read_pickle('raw_data/events.pickle')
events['repo_full_name'] = events.owner + '/' + events.repo
events['user_id'] = events.user_id.apply(md5_hash)
events['repo_full_name'] = events.repo_full_name.apply(md5_hash)
events.drop(columns=['owner', 'repo'])
events.to_pickle('data/events_anonymized.pickle')
events
```

```python
orgunits = pd.read_csv('raw_data/teams_extended.csv')
orgunits['partition'] = pd.to_datetime(orgunits.partition_date).dt.date
orgunits.id = orgunits.id.apply(md5_hash)
squads = orgunits[orgunits.type == 'squad'][['userid', 'partition', 'id']]
squads.to_pickle('raw_data/squads_anonymized.pickle')
squads
```
